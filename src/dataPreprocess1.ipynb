```python
import pandas as pd
```

```python
df=pd.read_csv("KarnalOriginal.csv")
```

```python
df.info()
```

```python
df.sample(2)
```

```python
all_dates = pd.date_range(
    start="1975-1-1",
    end="2025-12-31",
    freq='D'
)
```

```python
df['YEAR'] = pd.to_datetime(
    df['YEAR'].astype(str) + '-' +
    df['MN'].astype(str) + '-' +
    df['DT'].astype(str)
)
```

```python
df.sample(3)
```

```python
df = df.rename(columns={'YEAR': 'TIME'})
```

```python
df=df.drop(["MN","DT","INDEX"],axis=1)
```

```python
df = df.set_index('TIME')

df = df.reindex(all_dates)
df.index.name = 'TIME'
```

```python
df =df.reset_index()
```

```python
df.info()
```

```python
df.to_csv("Karnal.csv",index=False)
```
