```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("KarnalNew.csv")
df["TIME"]=pd.to_datetime(df["TIME"])
```

```python
df.info()
```

```python
temp=df[["TIME","RF"]]
temp=temp[temp["RF"].isna()]
temp["TIME"].dt.year.value_counts().sort_index().values[:11].sum()/365/51
```

```python
df['month'] = df['TIME'].dt.month
df['day'] = df['TIME'].dt.day
```

```python
df['RF'] = df['RF'].fillna(
    df.groupby(['month', 'day'])['RF'].transform('median')
)
```

```python
df['MIN'] = df['MIN'].fillna(
    df.groupby(['month', 'day'])['MIN'].transform('mean')
)
```

```python
df['MAX'] = df['MAX'].fillna(
    df.groupby(['month', 'day'])['MAX'].transform('mean')
)
```

```python
df=df.drop(["month","day"],axis=1)
```

```python
df.to_csv("Karnal.csv",index=False)
```
